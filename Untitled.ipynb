```python
def powOf(n,i):
    return (n & (1<<i))

def bitOf(n,i):
    return (powOf(n,i)>>i)

def DfsPostOrderIndex(n, N):
    n = n % 2**(N-1)
    return sum( [(2*powOf(n,i) - 1) for i in range(0,2**N) if bitOf(n,i)] ) #% ((2**N))

N = 8
order = (2**N) * ((2**N) - 1)
for n in range(0,order):
    d = DfsPostOrderIndex(n,N)
    #print("%4d: %4d" % (n,d))
```
